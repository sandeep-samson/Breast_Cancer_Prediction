# Breast Cancer Prediction

## Import Libraries

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import warnings
warnings.filterwarnings('ignore')
```

## Dowload dataset from Kaggle

```python
path = kagglehub.dataset_download('uciml/breast-cancer-wisconsin-data', 'data.csv')
```

## Load and Explore Data

```python
# load data on dataframe
df = pd.read_csv(path)
```

```python
# display dataframe
df.head()
```

```python
# count of rows and columns
df.shape
```

```python
# count number of null(empty) values
df.isnull().sum()
```

```python
# Drop the column with null values
df.dropna(axis=1, inplace=True)
```

```python
# count of rows and columns
df.shape
```

```python
# Get count of number of M and B cells in diagnosis
df['diagnosis'].value_counts()
```

## Label Encoding

```python
# Get Datatypes of each column in our dataset
df.dtypes
```

```python
# Encode the disgnosis values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1] = labelencoder.fit_transform(df.iloc[:,1].values)
```

```python
# display df
df
```

## Split Dataset & Feature Scaling

```python
# Splitting the dataset into independent and dependent datasets
X = df.iloc[:, 2:].values
Y = df.iloc[:, 1].values
```

```python
# Splitting datasets into training(75%) and testing(25%)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
# Scaling the data(feature scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# print data
X_train
```

## Build a Logistic Regression Model

```python
# Build a logistic regression classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
classifier.fit(X_train, Y_train)
```

```python
# make use of trained model to make prediction on test data
predictions = classifier.predict(X_test)
```

## Performance Evaluation

```python
# plot confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, predictions)
print(cm)
sns.heatmap(cm, annot=True)
```

```python
# Get accuracy score for model
from sklearn.metrics import accuracy_score
print("Model's Accuracy Score:", accuracy_score(Y_test, predictions))
```
